```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.shape
```

```python
 df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1, inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_Gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sns.barplot(x= 'Gender',y= 'Amount',data = sales_Gen)
```

```python
ax = sns.countplot(data = df,x = 'Age Group',hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_age,x= 'Age Group',y = 'Amount')
```

```python
sales_state = df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x= 'State',y = 'Orders')
```

```python
sales_A_state = df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_A_state,x= 'State',y = 'Amount')
```

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_mar = df.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

ax = sns.barplot(data = sales_mar,x = 'Marital_Status',y= 'Amount',hue = 'Gender')

sns.set(rc={'figure.figsize':(5,5)})
```

```python
ax = sns.countplot(data = df, x= 'Occupation')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_occ,x= 'Occupation',y = 'Amount')
```

```python
ax = sns.countplot(data = df, x= 'Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_pro = df.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pro,x= 'Product_Category',y = 'Amount')
```
